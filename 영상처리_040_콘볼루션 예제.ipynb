```python
# 합성곱 예제 
import numpy as np
from matplotlib import pyplot as plt

img = plt.imread('messi5.jpg')
img = np.sum( img*[.2126, .7152, .0722], axis=2 ).astype('uint8')

from scipy import signal
a = signal.convolve( img, [ [-1,-1,-1], [-1, 8,-1], [-1,-1,-1] ] )

plt.imshow( a, cmap='gray')
plt.show()
```
