```python
#  ,  ,     
import numpy as np
from matplotlib import pyplot as plt
#    
img = plt.imread( 'messi5.jpg' )

#     
red  = img[:, :, 0]  # blue channel data
green = img[:, :, 1]  # green channel data
blue   = img[:, :, 2]  # red channel data

#    
idx = 0 
fig, axis = plt.subplots( 1, 3, figsize=(18, 17) )

#   
average_gray =  red/3 + green/3 + blue/3 

ax = axis[ idx ]
ax.imshow( average_gray, cmap='gray' )
ax.set_title( 'Average' )
idx += 1 

# Luminosity  
luminosity_gray =  0.21*red + 0.72*green + 0.07*blue 

ax = axis[ idx ]
ax.imshow( luminosity_gray, cmap='gray' )
ax.set_title( 'Luminosity' )
idx += 1 

# Lightness  
lightness_gray = np.min( img, axis=2 )/2 + np.max( img, axis=2 )/2

ax = axis[ idx ]
ax.imshow( lightness_gray, cmap='gray' )
ax.set_title( 'Lightness' )
idx += 1 

#   
plt.show()
```
